```python
from capsa_tk.exchanges import Binance
import pandas as pd
from tqdm import tqdm
import requests
import xml
```

```python
binance = Binance()
```

```python
# Get tickers
tickers = []
with open("symbols_binance.txt") as f:
    contents = f.readlines()
for line in contents:
    if "USDT" in line and not "UP" in line and not "DOWN" in line:
        ticker = line.strip()[:-1].split("U")[0]
        ticker += "/USDT"
        tickers.append(ticker)
print(tickers)
```

```python
init = binance.get_ochlv(tickers[0], "1d", from_date="2021-01-01T00:00:00.000", to_date="2023-01-01T00:00:00.000")
init = init.set_index("date")
init = init.drop(["open", "high", "low",], axis=1)
df = init.rename(columns={"close": tickers[0], "volume": f"{tickers[0]}_volume"})

for ticker in tqdm(tickers[1:]):

    try:
        data = binance.get_ochlv(ticker, "1d", from_date="2021-01-01T00:00:00.000", to_date="2023-01-01T00:00:00.000").set_index("date")
        df[[ticker, f"{ticker}_volume"]] = data[["close", "volume"]]
    except:
       pass
df
```

```python
# Drop empty cols:
df = df.dropna(axis=1, how="all")
df.to_csv("data.csv", index=True)
```
