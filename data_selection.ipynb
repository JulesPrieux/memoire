```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data["date"] = pd.to_datetime(data["date"])
data = data.set_index("date")
data
```

```python
columns = list(data.columns)
tickers = [ticker for ticker in columns if not "volume" in ticker]
tickers_volume = [f"{ticker}_volume" for ticker in tickers]

# Get volume in $
data[tickers_volume] = data[tickers].values * data[tickers_volume].values
data
```

```python
average_daily_volume = data[tickers_volume].mean(axis=1)
plt.plot(average_daily_volume)
plt.gcf().autofmt_xdate()
plt.title("Average Daily traded volume ($)")
```

```python
# Keep only the pairs with enought volume.
def get_first_occurence_volume(ticker_volume: str) -> float:
    return data[ticker_volume].dropna().iloc[0]

to_drop = []
VOLUME_THRESHOLD = 10_000_000
for ticker in tickers_volume:
    if get_first_occurence_volume(ticker) < VOLUME_THRESHOLD:
        to_drop.append(ticker)
to_drop
df = data.drop(to_drop, axis=1)
display(df)
```

```python
df.to_csv("clean_data.csv")
```
